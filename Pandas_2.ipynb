#8. 결측치 처리 (1)

```python
import pandas as pd
import numpy as np

# 실습 데이터 생성
df = pd.DataFrame(data = np.arange(18).reshape(6,3),
				  index = ['a','b','c','d','e','f'],
                  columns=['col1','col2','col3'])
                  
df['col4'] = pd.Series(data = [1.7, 1.2, 2.4], 
                       index = ['a','e','c'])
df.loc['c'] = None
print('Sample Data')
print(df)

# info()
print(df.info())
# 해석 : non-null:정상값 , 5:정상값의 개수 
# index = 6 -> col(1~4) 당 행의 개수가 6개인데 이중에서 5개는 정상값이기에 자동적으로 1개가 결측값임을 알수 있다
print()

# isnull(), isna() - 정상값과 결측값의 결과를 Treu, False 로 나타내는 함수
print(df.isnull()) # null 값이 True!

# isnull().sum() - 실제로 데이터수가 굉장히 많기 때문에  가장 많이 사용되는 방식 
# 각 컬럼 별 결측값의 개수 
print(df.isnull().sum())
# 각 행별로 결측값 개수도 가능 axis=1 로 사용
print(df.isnull().sum(axis=1))
```

#8.1 결측치 처리(2)

```python
import pandas as pd
import numpy as np

# 데이터의 결측치를 찾아서 처리하는 함수 

# 실습 데이터 생성
df = pd.DataFrame(data = np.arange(18).reshape(6,3),
				  index = ['a','b','c','d','e','f'],
                  columns=['col1','col2','col3'])
                  
df['col4'] = pd.Series(data = [1.7, 1.2, 2.4], 
                       index = ['a','e','c'])
df.loc['c'] = None
print('Sample Data')
print(df)

print()

# 1. dropna() 함수 - 결측값이 포함되어있는 행을 모두 삭제하고 남아있는 데이터 값을 반환
# print(df.dropna())
print()
# 이렇게 모두 삭제해버리면 데이터 분석도 하기 전에 데이터의 크기가 너무 작아짐 해결책으로
# how='all' 사용 - 행이 all 결측값인 행만 삭제
print("d",df.dropna(how='all',inplace=True))
print()

# 2. 열을 기준으로 결측치 조회하여 있으면 제거 
# axis = 1 or axis = columns 둘중 아무거나 
print(df.dropna(axis = 'columns')) 

#3. 다시 결측값으로 만들기
df.iloc[:2 ,0] =np.nan
df.iloc[:4,1]=np.nan
print(df)

#4. fillna() 결측값 변환하기 
print(df.fillna(0)) 
# 이때 ()안에는 int, float, str, dic, series, dataframe 모두 가능

#5. dic 로 변환하기
replace  = {'col1':0,'col2':100}
print(df.fillna(replace))

# 6.  대표적인 평균을 활용하여 결측지 대체! 하기 
# 원본데이터 유지하기 
function = {'col1':df['col1'].mean()}
print(df.fillna(function, inplace =True))
print(df)
```

#9. 데이터 조회를 위한 인덱싱

```python
import pandas as pd

# 실습 데이터 생성
d = {'name': ['Jessi', 'Emma', 'Alex', 'Jessi', 'Tom'],
     'score': [100, 95, 80, 85, 97],
     'grade': ['A', 'A', 'B', 'B', 'A'],
     'subject':['python', 'java', 'python', 'c', 'java']}

df = pd.DataFrame(data=d)
print('Sample Data')
print(df)

#1. set_index() : 컬럼 값들을 행인덱스로 바꾸기
# index 의 값을 사라지고 컬럼이 그 자리로 배치됨 
df.set_index('name', inplace = True)
print(df)

print()

# loc[] - 인덱스 명(값) 을 기준으로 행 데이터 조회 - 문자 
list_loc = df.loc[['Jessi','Emma']]
print(list_loc)

# iloc[] - 인덱스 번호를 기준으로 행 데이터 조회 - interger 
list_iloc = df.iloc[[0,4]]
print(list_iloc)

print()
# loc[] 활용- boolean indexing 
booleaning = df.loc[df['score']>=95]
print(booleaning)

# loc[]활용 - 2
booleaning2 = df.loc[df['subject']=='python']
print(booleaning2)
```

#11. Pandas 의 집계함수

```python
import pandas as pd
import numpy as np

# 실습 데이터 생성
score = {'sub1': [3, 9, 1, 1, 9],
         'sub2': [2, 9, np.nan, np.nan, 8],
         'sub3': [np.nan, 1, 5, 5, 7],
         'sub4': [np.nan, 3, np.nan, 1, np.nan]}

df = pd.DataFrame(data=score)
print(df)

# Numpy 는 axis=0,1,2 모두 적용 가능하지만 
# Pandas 는 axis = 0,1 만 가능하고 굳이 나와있지 않다면 0 을 기본으 호출한다

#1. count() - sub을 기준으로 데이터의 수 즉, 행 (axix=0) 기준
print(df.count()) 
print()

#2. count(axis=1)
print(df.count(axis=1))
print()

#3. sum() - axis=0 기준 
print(df.sum())
print()

#4. sum(skipna=False) 기본 집계함수는 결측값을 뺴고계산하는데 결측값을 포함해서 계산하고 싶을때 
print(df.sum(skipna=False))
# 하나라도 계측값이 존재하면 결과는 계측값으로 나온다 
```

#12.데이터 그룹화 

```python
import pandas as pd
import numpy as np

# 실습 데이터 생성
sample = {'product':['a','b','a','b','a','b','a','a'],
          'sensor':['s1','s1','s2','s3','s2','s2','s1','s3'],
          'x':np.arange(1,9),
          'y':np.arange(5,13)}

df = pd.DataFrame(data=sample)
print(df)

# 단계 0. key, value 의 값을 가지고 있는 input
#      1. 중심이 되어 묶일 데이터를 사용해 groupby = ('묶일 데이터 ') => split
#      2. 새롭게 묶인 데이터를 가지고 ex> Mean() 적용 => apply
#      3. 나온 값을 가지고 다시 그룹화 => combine 
# 모든 단계에서의 data 는 key, value 로 이루어져 있다! 

print()

#1. 그룹화하기
list_ = df.groupby('product')
print(list_)
print()

#2. for 문을 이용해서 key, value 값 확인
for key, value in list_:
  print(key)
  print(value)
  # 보면 product 기준으로 a,b 로 나뉜것을 볼 수 있음

print()

#3. 집계함수 적용하기
print("-------",list_.sum())
# 이렇게도 가능 print(df.groupby('product'),sum()) 
# pandas 에는 굳이 표시 없으면 axis =0 으로 sum 이 이루어짐
print()

#4. 2가지로 그룹화하기
list2 = df.groupby(['product','sensor']).sum() # 2중으로 묶을때는 안에 목록 - 리스트로 대입
print(list2)
# 첫번째로 product 로 그룹화 -> 두번째로 sensor 로 그룹화 -> data 로 sum

print()

#5. 그룹화 한 뒤 'x' 로만 다시 한번 그룹화 - 중첩
list3 = df.groupby(['product','sensor'])['x'].sum() #구문 조심하기! 
print(list3)
print()

#6. 그룹화 한뒤 각각의 그룹에 각기 다른 집계함수 적용하기
function= {'x':'max', 'y':'min'} # 조건은 딕셔너리로 정의 : 적용함수를 딕셔너리로 정의 - 중요!!
list4 =  df.groupby(['product', 'sensor']).agg(function)
# agg(x) ->? : 조건을 인자로 지정해주는 코딩 
# 즉 x 변수를 딕셔너리로 생성 후 agg 함수를 통해 적용해주는것 -> 방식 잘 생각해서 쓰기!!  
print(list4)
```

#13. 데이터 결합하기

```python
import pandas as pd

# 실습 데이터 생성
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])                   

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])

df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                    index=[2, 3, 6, 7])
                    
print('DF1')
print(df1)
print('\nDF2')
print(df2)
print('\nDF3')
print(df3)
print('\nDF4')
print(df4)

print()

#1. df1 , df2 데이터 결합 - shape 이 일치할 떄
result =  pd.concat([df2,df1,df3],ignore_index = True) 
print(result)
# 명시되어 있지 않으면 axis=0을 기준으로 결합되기에
# 컬럼은 일치하되 행의 수가 늘어난다 
# 순서도 중요 []에 쓴 대로 결합이 된다. 
# ingore_index = true : 데이터가 결합될 떄 기존의 index 를 무시하고 0부터 새로운 index를 주기 위한 값


print()

# 2. shape 이 서로 다른 데이터 결합 -> 행을 기준으로 결합후, 채울 수 없는 곳은 결측값으로 대체한다
result = pd.concat([df1, df4])
print(result)
print()

# 3. 슬라이싱을 통해 일부 데이터 합치기
result = pd.concat([df1[:2], df4[3:]])
print(result)
print()

# 4. axis =1 을 기준으로 데이터 결합하기
result = pd.concat([df1,df4],axis=1,ignore_index= True)
# axis =1 을 기준으로 결합시키면 (4,4),(4,3) = 이때 행으로 늘어나느것이 아니라 컬럼수가 증가 - 7개
# but 행도 df1 과 다른 df4 에 있는 행은(6,7) 새로 추가됨 
# ignore_index=True : 컬럼값이 중복인 값이 있는데 그것을 무시하고 새로운 index를 0부터 새로운 컬럼값으로 대입
print("4",result)
print()

# 5. 교집합으로 데이터 결합하기 join = 'inner'
# join을 넣지 않으면 합집합 기준으로 데이터를 결합
# axis=0 으로 분석을 하면 겹치는 컬럼이 존재하고
# axis=1 로 분석을 하면 겹치는 행의 값이 존재하기에
# 어떤 axis 르 쓰는지에 따라 다른 값이 결정되어 교집합이 되어 결합됨! 
result = pd.concat([df1,df4],join='inner')
print(result)
```
